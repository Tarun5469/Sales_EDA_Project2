```python
import matplotlib.pyplot as plt
Days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
storeA=[250,300,400,350,380,450,500]
storeB=[300,450,450,650,609,565,800]
plt.plot(Days,storeA,linewidth=1)
plt.plot(storeB,linewidth=1)
plt.xlabel("Sales at store_A",fontsize=15,color="r")
plt.ylabel("Sales at store_B",fontsize=15,color="g")
plt.title("SALES REPORT",fontsize=15,fontname="Brush Script MT")
plt.grid(True)
#plt.style.use('Solarize_Light2') 
plt.show()
```
